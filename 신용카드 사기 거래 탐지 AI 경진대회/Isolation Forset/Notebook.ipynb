```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

# 스케일링 하기 전 모든 데이터를 합침
trainnum = len(train)
valnum = len(val)
testnum = len(test)

all_data = pd.concat([train, val, test])
all_data = all_data.drop('Class', axis=1)
all_data
```

```python
all_data_scaled = scaler.fit_transform(all_data)
all_data_scaled = pd.DataFrame(all_data_scaled, columns=cols, index=all_data.index)
all_data_scaled
```

```python
train_scaled = all_data_scaled[:trainnum]
val_scaled = all_data_scaled[trainnum:trainnum+valnum]
test_scaled = all_data_scaled[trainnum+valnum:]
```

```python
train_scaled.hist(bins=50, figsize=(20, 20))
plt.show()
```

```python
val_normal, val_fraud = val['Class'].value_counts()
val_contamination = val_fraud / val_normal
val_contamination
```

```python
from sklearn.metrics import f1_score
from sklearn.ensemble import IsolationForest
from sklearn.metrics import classification_report

import warnings
warnings.filterwarnings(action='ignore')
```

```python
IForest = IsolationForest(n_estimators=125, # 생성할 의사결정 트리 개수 (한 예측기에 몇번의 시행을 할 것인가)
                          max_samples=len(train_scaled), # 한 예측기에 들어가는 샘플의 비율
                          contamination=val_contamination, # 불량의 비율
                          verbose=0)
IForest.fit(train_scaled)
```

```python
def get_pred_label(model_pred):
    # IsolationForest는 1:정상 -1:불량 이므로 문제에 맞게 0정상 1불량으로 레이블 변환
    model_pred = np.where(model_pred == 1, 0, model_pred)
    model_pred = np.where(model_pred == -1, 1, model_pred)
    return model_pred
```

```python
val_x = val_scaled
val_y = val['Class']

val_pred = IForest.predict(val_x)
val_pred = get_pred_label(val_pred)
val_score = f1_score(val_y, val_pred, average='macro')
val_score
```

```python
%%time

test_pred = IForest.predict(test_scaled)
test_pred = get_pred_label(test_pred)
```

```python
submit = pd.DataFrame()

submit['ID'] = submission.index
submit['Class'] = test_pred
submit.to_csv('submit.csv', index=False)
submit
```
